**Data Visualisation**

```python
#Pie Chart for Song Genre
count=song['genre'].value_counts()
plt.pie(count,labels=count.index,autopct='%1.1f')
plt.title('Song Genre')
plt.show()
```

```python
#Bar plot between top 10 Artists and their  mean popularity
artist_popularity = song.groupby('artist_name')['popularity'].mean().sort_values(ascending=False)

top_10_artists = song['artist_name'].value_counts().head(10).index

top_artist_popularity = artist_popularity.loc[top_10_artists]

plt.figure(figsize=(12, 6))
sns.barplot(x=top_artist_popularity.index, y=top_artist_popularity.values, palette='viridis')
plt.title('Mean Popularity of Top 10 Artists')
plt.xlabel('Artist Name')
plt.ylabel('Mean Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Scatter plot between Acousticness and Popularity
plt.scatter(song['acousticness'],song['popularity'])
plt.title('Acousticness vs Popularity')
plt.xlabel('Acousticness')
plt.ylabel('Popularity')
plt.show()
```

```python
#Box Plot between Instrumentalness and Key
plt.figure(figsize=(12, 6))
sns.boxplot(x='key', y='instrumentalness', data=song, palette='plasma')
plt.title('Instrumentalness vs. Key')
plt.xlabel('Key')
plt.ylabel('Instrumentalness')
plt.show()
```

```python
#Pie Chart to visualise unique keys
count=song['key'].value_counts()
plt.pie(count,labels=count.index,autopct='%1.1f')
plt.title('Song Key')
plt.axis('equal')
plt.show()
```

```python
#Line chart between liveness and tempo
sort_liveness=song.sort_values('liveness',ascending=False)
sort_tempo=song.sort_values('tempo',ascending=False)

plt.plot(sort_liveness['liveness'],sort_tempo['tempo'],color='red')
plt.title('Liveness vs Tempo')
plt.xlabel('Liveness')
plt.ylabel('Tempo')
plt.grid()
plt.show()
```

```python
#Count Plot to visualise the signing date of songs
count=song['time_signature'].value_counts()
plt.bar(count.index, count.values,color=['pink','aqua','blue','skyblue'])
plt.title('Song Time Signature')
plt.show()
```

```python
#Pie chart to represent various modes of songs
count=song['mode'].value_counts()
plt.pie(count,labels=count.index,autopct='%1.1f')
plt.title('Song Mode')
plt.show()
```

```python
#Heatmap to correlate the data in the different columns
numeric_cols = song.select_dtypes(include=np.number).columns
correlation_matrix = song[numeric_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
#Histogram to show the distribution in Danceability
plt.hist(song['danceability'],bins=10,color='lightgreen')
plt.title('Danceability Distribution')
plt.xlabel('Danceability')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
